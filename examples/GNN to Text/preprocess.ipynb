```python
import pandas as pd

root_path = "../uploads/ode-gnn"
drug_gene = pd.read_csv(f"{root_path}/DGIdb_interactions.tsv", sep="\t")
drug_gene
```

```python
clin = pd.read_excel(f"{root_path}/rna_data.xlsx", sheet_name="PCT raw data")
clin
```

```python
disease_gene = pd.read_csv(f"{root_path}/BRCA_GENE_OT.tsv", sep="\t")
disease_gene
```

```python
print(clin["Tumor Type"].unique())
print(clin["Treatment"].unique())
```

```python
def normalize(name):
    t = name.lower()
    t = t.replace("abraxane", "nab-paclitaxel").replace("5fu", "5-fu")
    t = t.replace("-3.5mpk", "").replace("-50mpk", "").replace('"', "")
    return t


drug_names = set()
for name in sorted(clin["Treatment"].unique()):
    if "+" in name:
        for part in name.split("+"):
            drug_names.add(part.strip())
    else:
        drug_names.add(name.strip())

drug_gene_dict = {}
for name in sorted(drug_names):
    if "+" in name:
        continue
    t = normalize(name)
    df = drug_gene[drug_gene["drug_claim_name"].str.lower() == t.lower()]
    df = df.dropna(subset=["gene_name"])
    drug_gene_dict[name] = set(df["gene_name"].unique())
    print(f"Treatment: {name}, Number of associated genes: {len(drug_gene_dict[name])}")
for name in sorted(clin["Treatment"].unique()):
    if "+" in name:
        drug_gene_dict[name] = set()
        for part in name.split("+"):
            drug_gene_dict[name].update(drug_gene_dict[part.strip()])
        print(f"Treatment: {name}, Number of associated genes: {len(drug_gene_dict[name])}")

flat = []
for name, genes in sorted(drug_gene_dict.items()):
    for gene in sorted(genes):
        flat.append({"Treatment": name, "Gene": gene})
drug_gene_df = pd.DataFrame(flat)
drug_gene_df.to_csv(f"{root_path}/drug_gene.csv", index=False)
```
